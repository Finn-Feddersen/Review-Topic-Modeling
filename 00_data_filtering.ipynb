```python
# import necessary packages
import pandas as pd
import numpy as np
import json
```

```python
# read in data from Cell_Phones_&_Accessories.json
data = []
with open('Cell_Phones_and_Accessories.json', 'r') as file:
    for line in file:
        data.append(json.loads(line))

df = pd.DataFrame(data)
```

```python
# read in meta-data from meta_Cell_Phones_and_Accessories.json
data = []
with open('meta_Cell_Phones_and_Accessories.json', 'r') as file:
    for line in file:
        data.append(json.loads(line))

df_meta = pd.DataFrame(data)
```

```python
df_meta.head()
```

```python
# filter for apple df_meta and extract the asin numbers (product IDs)
df_meta = df_meta[df_meta['brand'] == 'Apple']
df_meta = df_meta[['asin', 'brand']]
print(df_meta.shape)
df_meta.head()
```

```python
# merge df and df_meta to filter for only apple products (based on the product IDs)
df = pd.merge(df, df_meta, on='asin', how='inner')
print(df.shape)
df.head()
```

```python
# write (much smaller) df to csv file
df.to_csv('apple.csv', index=False)
```
